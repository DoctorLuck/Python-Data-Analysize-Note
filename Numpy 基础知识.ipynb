#### 利用数组进行数据处理

```python
points = np.arange(-5,5,0.01)
points
```

```python
# np.meshgrid()接受两个一维数组，并产生两个二维矩阵
xs,ys = np.meshgrid(points,points)
xs
```

```python
import matplotlib.pyplot as plt
z = np.sqrt(xs**2 + ys**2)
z
```

```python
plt.imshow(z,cmap=plt.cm.gray)
plt.show()
```

#### 将条件逻辑表述为数组运算

np.where()是三元表达式x if condition else y的矢量化版本

```python
xarr = np.array([1,3,5,7])
yarr = np.array([2,4,6,8])
cond = np.array([True,False,True,True])
np.where(cond,xarr,yarr)
```

np.where()的第二个和第三个参数不必是数组，都可以是标量值

### 用于数组的文件输入及输出

np.save()  np.load()  默认情况下，数组是以未压缩的原始二进制格式保存在扩展名为.npy的文件中的

```python
arr = np.arange(10)
arr
```

```python
np.save('some_array',arr)
# 文件路径末尾没有扩展名.npy  会自动加上
```

```python
np.load('some_array.npy')
```
